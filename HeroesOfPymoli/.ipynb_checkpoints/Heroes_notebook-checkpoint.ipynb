```python
import pandas as pd
purchase_data = "Resources/04-Pandas_Homework_HeroesOfPymoli_Resources_purchase_data.csv"
purchase_df = pd.read_csv(purchase_data)
purchase_df.head(10)
```

```python
players = purchase_df["SN"].unique()
Total_Players = pd.DataFrame({"Total Players": [players.size]})
Total_Players
```

```python
items = purchase_df["Item Name"].unique()
price = purchase_df["Price"].values
average_price = purchase_df["Price"].mean()
Summary = pd.DataFrame({"Number of Unique Items": [len(items)], "Average Price": [average_price], "Number of Purchases": [len(purchase_df)], "Total Revenue": [sum(price)]})
Summary
```

```python
Gender = pd.DataFrame(purchase_df, columns=["SN", "Gender"])
Gender = Gender.drop_duplicates()
Gender_count = Gender["Gender"].value_counts()

Gender_count_total = pd.DataFrame({"Total Count": Gender_count})
total_users = Gender_count_total["Total Count"].sum()
Gender_count_total["Percentage of Players"] = Gender_count_total["Total Count"]*100/total_users
Gender_count_total
```

```python
Grouped_gender = purchase_df.groupby("Gender").agg({
    "Purchase ID": "count",
    "Price": ["mean", "sum"]
})
Grouped_gender["", "Avg Total Purchase per Person"] = Grouped_gender[("Price", "sum")]/Gender_count_total["Total Count"]
renamed_df = Grouped_gender.rename(columns={"sum":"Total Purchase Value", "mean":"Average Purchase Price", "count":"Purchase Count", "Purchase ID":"", "Price":""})
renamed_df.head()
```

```python
# establish age bins (<10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, >40)
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_df["Age ranges"] = pd.cut(purchase_df["Age"], bins, labels=group_names)
# purchase_df.head()
# groupby
bins_df = purchase_df.groupby("Age ranges").agg({
    "Purchase ID": "count"
})
bins_df
# unique users only
```

```python
bins2_df = purchase_df.groupby("Age ranges").agg({
    "Purchase ID": "count",
    "Price": ["mean", "sum"]
})
bins2_df
# add column with Avg per unique person
```

```python
Grouped_SN = purchase_df.groupby("SN").agg({
    "Purchase ID": "count",
    "Price": ["mean", "sum"]
})
Sorted_SN = Grouped_SN.sort_values(("Price", "sum"), ascending=False)
Sorted_SN.head()
# rename columns (ran out of time)
```
